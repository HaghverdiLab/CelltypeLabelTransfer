```r
library(ggplot2)
get_combinations <- function(files, size, seed){
    set.seed(seed)
    return(sample(files, size = size))
}

translate <- function(vector){
    vector <- stringr::str_replace_all(vector, "\\.\\.", "+ ")
    vector <- stringr::str_replace_all(vector, "\\.", " ")
    return(vector)
}


options(repr.plot.width=20, repr.plot.height=10)

```

```r
tags_all <- list.files("../Data/Predictions/ItClust/", pattern = "PBMC10x")
#tags <- tags[stringr::str_detect(tags,  "_100_")]
length(tags_all)
```

```r
#tags <- paste(sep="_", "PBMC10x", c(38,100,250,1000), seq(1,20,1))
x <- expand.grid(list("PBMC10x", c(38,100,250,1000), seq(1,20,1)))
tags <- apply(x, 1, function(row) paste(collapse ="_", unlist(row), sep=""))
tags <- stringr::str_replace_all(tags, " ", "")
tags <- tags[tags %in% tags_all]
```

```r
readSeurat <- function(folder, tag){
    file <- paste0(paste(sep="/",folder, tag), "_predictions.txt")
    
    seurat <- read.csv(file, sep="\t")
    seurat$id <- rownames(seurat)
    seurat <- seurat[, c(c("predicted.id", "prediction.score.max",
                                         "class_", "prediction.match", "id"))]
    seurat <- reshape2::melt(seurat,id=c("predicted.id", "prediction.score.max",
                                         "class_", "prediction.match", "id"),
                            value.name="confidence_score", variable.name="celltype")
    seurat$celltype <- stringr::str_replace_all(seurat$celltype, "prediction.score.","")
    seurat$celltype <- translate(seurat$celltype)
    seurat <- seurat[,c("id", "class_", "predicted.id", "celltype",
                        "prediction.match", "confidence_score")]
    colnames(seurat) <- c("id", "correct_label", "predicted_label",
                          "celltype", "correct_predicted", "confidence")
    
    seurat <- adjust_df(seurat, tag, "Seurat")
    return(seurat)
}
adjust_df <- function(data, tag, method){
    data$is_correct <- data$correct_label == data$celltype
    data$was_predicted <- data$predicted_label == data$celltype
    data$method = method
    x <-  stringr::str_split_fixed(tag, '_', 3)  
    data$reference <- x[1]
    data$size <- x[2]
    data$set <- x[3]
    return(data)
}
seurat<- do.call(rbind,lapply(tags, function(tag) readSeurat("../Data/Predictions/Seurat/", tag)))
head(seurat)
options(repr.plot.width=20, repr.plot.height=10)

ggplot(seurat, aes(celltype, confidence, color=is_correct)) + 
geom_boxplot(alpha=0.3, width=0.1, position = position_dodge(0.5))+ geom_violin(position=position_dodge(0.5), alpha=0)+
facet_grid(rows=vars(size))+theme_bw()+
theme(legend.position = "bottom")                                 
```

```r
readSCN <- function(folder, tag){
    scn <- read.csv(paste0(paste(sep="/",folder, tag), ".txt"), sep="\t")
    rownames(scn) <- scn$id 
    scn$predicted[is.na(scn$predicted)]<- "rand"
    scn <- scn[!is.na(scn$class_),] 
    scn <- scn[!is.na(scn$prediction.match),]
    scn <- reshape2::melt(scn,id=c("predicted", "nGene", "nUMI", "percent.mito", "class_",
                                   "Cluster", "Experiment", "Method","prediction.match", "id"),
                            value.name="confidence_score", variable.name="celltype")
    scn <- scn[,c("id", "class_", "predicted", "celltype","prediction.match", "confidence_score")]
    colnames(scn) <- c("id", "correct_label", "predicted_label", "celltype", "correct_predicted", "confidence")
    
 
    scn$celltype <- stringr::str_replace_all(scn$celltype, "\\.\\.", "+ ")
    scn$celltype <- stringr::str_replace_all(scn$celltype, "\\.", " ")

    scn <- adjust_df(scn, tag, "SCN")   
    return(as.data.frame(scn))
}

scn<- do.call(rbind,lapply(tags, function(tag) readSCN("../Data/Predictions/SCN/", tag)))
head(scn) 
                           
ggplot(scn, aes(celltype, confidence, color=is_correct)) + 
geom_boxplot(alpha=0.3, width=0.1, position = position_dodge(0.5))+ geom_violin(position=position_dodge(0.5), alpha=0)+
facet_grid(rows=vars(size))+theme_bw()+
theme(legend.position = "bottom")                    
                                
```

```r
readSingleR <- function(folder, tag){
    singler <- read.csv(paste0(paste(sep="/",folder, tag), ".txt"), sep="\t")

    singler <- reshape2::melt(singler,id=c("predicted", "nGene", "nUMI", "percent.mito", "class_",
                                   "Cluster", "Experiment", "Method","prediction.match", "id",
                                  "first.labels", "tuning.scores.first", "tuning.scores.second", "pruned.labels"),
                            value.name="confidence_score", variable.name="celltype")
    singler <- singler[,c("id", "class_", "predicted", "celltype","prediction.match", "confidence_score")]
    colnames(singler) <- c("id", "correct_label", "predicted_label", "celltype", "correct_predicted", "confidence")
    singler$celltype <- stringr::str_replace(singler$celltype, "scores.", "")
    singler$celltype <- translate(singler$celltype)
    singler <- adjust_df(singler, tag, "SingleR")   
                     
    return(singler)
}

singleR<- do.call(rbind,lapply(tags, function(tag) readSingleR("../Data/Predictions/SingleR", tag)))
head(singleR)
ggplot(singleR, aes(celltype, confidence, color=is_correct)) + 
geom_boxplot(alpha=0.3, width=0.1, position = position_dodge(0.5))+ geom_violin(position=position_dodge(0.5), alpha=0)+
facet_grid(rows=vars(size))+theme_bw()+
theme(legend.position = "bottom")                                                                        
```

```r
readItclust <- function(folder, tag, meta){
    file <- paste(sep="/", folder, tag)
    itclust <- read.csv(paste(sep="/", file, "results/clustering_prob.csv"))
    ids <- read.csv(paste(sep="/", file, "results/celltype_assignment.txt"), header=F)
    rownames(itclust) <- itclust$X
    itclust$X <- NULL
    ids[c("x", 'Cluster', 'x', "confidence", "x", "x", "type")] <-  stringr::str_split_fixed(ids$V1, ' ', 7)
    ids$type <- unlist(lapply(ids$type, function(type) substr(type, 1, nchar(type)-5)))
    ids$Cluster <- paste0("cluster", ids$Cluster)
    colnames(itclust) <- ids$type[ids$Cluster ==colnames(itclust)]
    colnames(itclust) <- translate(colnames(itclust))
    itclust$predicted <- unlist(apply(as.data.frame(itclust), 1, function(row) names(row)[which.max(row)]))
    itclust$id <- stringr::str_replace(rownames(itclust),"-target", "")
                                       
    itclust <- merge(itclust, meta)                          
    itclust$match <- itclust$class_ == itclust$predicted 
    itclust <- reshape2::melt(itclust,id=c("id","predicted","nGene", "nUMI", "percent.mito","Cluster","class_", "Experiment", "Method","match"),
                              value.name="confidence_score", variable.name="celltype")   
    
    itclust <- itclust[,c("id", "class_", "predicted", "celltype","match", "confidence_score")]
    colnames(itclust) <- c("id", "correct_label", "predicted_label", "celltype", "correct_predicted", "confidence")
 
    itclust <- adjust_df(itclust, tag, "ItClust")                                    
    print(paste(tag,nrow(itclust)))                                
    return(itclust)

}
meta <- read.csv("../Data/Fulldata/PBMC_Query//meta.csv") 
                                      
itclust <- do.call(rbind,lapply(tags, function(tag) readItclust("../Data/Predictions/ItClust", tag,meta)))
head(itclust )
ggplot(itclust, aes(celltype, confidence, color=is_correct)) + 
geom_boxplot(alpha=0.3, width=0.1, position = position_dodge(0.5))+ geom_violin(position=position_dodge(0.5), alpha=0)+
facet_grid(rows=vars(size))+theme_bw()+
theme(legend.position = "bottom")                                                                                                                    
```
