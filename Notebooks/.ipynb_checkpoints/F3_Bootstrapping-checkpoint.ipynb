```r
source("../Scripts/functions.R")
source("../Scripts/visulizations.r")
library(ggplot2)
library(dplyr)
library(purrr)
library(Seurat)
library(viridis)
library("RColorBrewer")
```

```r
celltypes = c("Cytotoxic T cell", "CD4+ T cell", "CD14+ monocyte", "B cell", "Megakaryocyte",
              "Natural killer cell",
              "CD16+ monocyte", "Dendritic cell", "Plasmacytoid dendritic cell")
methods <- c("Seurat",  "SingleR","CellID", "SingleCellNet", "ItClust")
sizes <- c(3090, 2418, 1373, 1022, 703, 623, 273, 126, 23)
names(sizes) <- celltypes
coi <- c("Cytotoxic T cell", "Natural killer cell","Dendritic cell", "Plasmacytoid dendritic cell" )

colors <-brewer.pal(length(celltypes), "Paired")
names(colors) <- celltypes
```

```r
umapdata <- read.csv("../Results/Files/umap_data.csv")
expressiondata_file="../Data/Fulldata//PBMC_Query/data.csv"
umap <- prepare_umap(file = expressiondata_file, meta_data = umapdata , split = "tech")
```

```r
plot_celltypes = plot_umap(umap, c("class"),NULL, colors,1,1, legend_ncol = 2)
options(repr.plot.width=10, repr.plot.height=6) 
f1a <- plot_celltypes[[1]]
f1a
```

```r


plots <- lapply(c("Seurat",  "SingleR", "CellID", "SCN", "ItClust"), function(method) plot_features(umap, method,
                                                                                        names("mono"),
                                                                                          sets=c("mono_dif"),
                                                                                          legend="left"))

f3a <-  ggpubr::ggarrange(plotlist=plots, nrow=5, ncol=1, common.legend = T)
options(repr.plot.width=5, repr.plot.height=15) 
#pred_plots <- lapply(plots, function(plot) plot$'Predicion-based' )                     
#pred_plot <- ggpubr::ggarrange(plotlist=pred_plots, nrow=4, ncol=1, common.legend = T)
f3a
```

```r
summary <- read.csv("../Results/Files/bootstrap_summary.csv")
summary_melt<- summary %>% 
           dplyr::group_by(method, class_, set) %>% 
dplyr::summarize(mono= mean(match_mono), mosaic = mean(match_mosaic)) 
head(summary_melt)
```

```r
full <- getSubset(data_indi_accuracy, 3090, "full")
head(full)
```

```r
ind <- getSubset(data_indi_accuracy, 100, "individual")
head(ind)
```

```r
data <- merge(full, ind, by=c("method", "class_"))
head(data)
```

```r
data_melt <- reshape2::melt(data, id=c("method", "class_", "set",  "full"),
                           value="accuracy", variable="approach")
head(data_melt)
```

```r
options(repr.plot.width=10, repr.plot.height=6) 
data_melt$class <- factor(data_melt$class_, levels=celltypes)
data_melt$method[data_melt$method=="SCN"] <- "SingleCellNet"
data_melt$method <- factor(data_melt$method, levels=methods)


get_accuracy_plot <- function(data){
  plot <- ggplot(data, aes(approach, value, color=approach, fill=approach))+
        geom_hline(aes(yintercept = full ), color="red", linetype="dashed")+
geom_violin(alpha=0.3) + 
geom_boxplot(width=0.1, alpha=0.3)+ 
facet_grid(rows=vars(method), cols=vars(class_), labeller = label_wrap_gen(width=12)) +

theme_bw() + labs(color="Cell type", y="Accuracy", fill="Cell type")+
theme(axis.text=element_text(size = 6), axis.title=element_text(size = 8),
      axis.text.x=element_text(angle = 10, hjust = 1), axis.title.x=element_blank(),
      strip.text = element_text(size = 6),
     legend.position="none")
return(plot)
}

accuracy_prediction <- get_accuracy_plot(data_melt[data_melt$class_ %in% coi,])
accuracy_prediction
```

```r
f3ab <- ggpubr::ggarrange(f3a, accuracy_prediction,
                          nrow=1, ncol=2, common.legend = F,
                          labels=c("A", "B"),widths=c(1,1.5)) 
f3ab
```
