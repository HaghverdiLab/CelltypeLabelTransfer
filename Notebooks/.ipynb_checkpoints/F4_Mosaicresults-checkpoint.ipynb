```r
celltypes = c("Cytotoxic T cell", "CD4+ T cell", "CD14+ monocyte", "B cell", "Megakaryocyte", "Natural killer cell",
              "CD16+ monocyte", "Dendritic cell", "Plasmacytoid dendritic cell")
methods <- c("Seurat",  "SingleR","CellID", "SCN", "ItClust")
sizes <- c(3090, 2418, 1373, 1022, 703, 623, 273, 126, 23)
names(sizes) <- celltypes

colors <-brewer.pal(length(celltypes), "Paired")
names(colors) <- celltypes
```

```r
get_datafull <- function(data, maxsize){
    data <- data[data$Size == maxsize,]
    data <- data[, !colnames(data) %in% c("Size", "Set", "refSize")]
    return(data)
}

get_accuracy_data <- function(file, methods, celltypes, type="mosaic"){
    data <- read.csv(file)
    data_accuracy <- calculate_accuracy(data)
    
    data_accuracy$Approach <- factor(data_accuracy$Approach, levels=methods)
    data_accuracy$class <- factor(data_accuracy$class, levels=celltypes)
    if(type== "mono") data_accuracy[, colnames(data_accuracy) != "Reference"]
    return(data_accuracy)
}
```

```r
data_mosaic      <- get_accuracy_data("../Results/Files/results_PBMCMosaic_long.csv",
                                      methods, celltypes, "mosaic")
```

```r
no_bootstrap <- read.csv("../Results/Files/results_PBMCMosaic_long.csv")
head(no_bootstrap)
no_bootstrap <- no_bootstrap[no_bootstrap$Size==102 & no_bootstrap$Reference=="PBMCMosaicBalanced",]
no_bootstrap_summary <- no_bootstrap %>% 
           dplyr::group_by(Approach, class, Set) %>% 
dplyr::summarize(accuracy= mean(Match)) 
individual_pred <- no_bootstrap_summary %>% 
  group_by(Approach, class) %>% 
  summarize(Q1=quantile(accuracy, probs = 0.25),
            Q3=quantile(accuracy, probs = 0.75),
            mean=mean(accuracy))
colnames(individual_pred) <- c("method", "class_", "Q1", "Q3", "mean")
head(individual_pred)
```

```r
data <- merge(summary_melt,individual_pred, by=c("method", "class_"))
names <- data.frame(names=as.factor(c("binary", "BinaryFiltered", "confidence", "confidenceFiltered")),
                  ids=as.factor(c("Prediction-based", "Prediction-based (Filtered)",
                          "Confidence score-based", "Confidence score-based (Filtered)")))
data$bootstrap_method <- names$ids[match(data$bootstrap_method, names$names) ]

head(data)
```

```r
data_bootstrap <- read.csv("../Results/Files/PBMC_bootstrap_mosaic.csv")
head(data_bootstrap)
```

```r
get_measures <- function(data, type, ref, method, size, set, col="Prediction"){
    data <- data[data$Reference == ref & data$Approach == method & data$Size == size & data$Set == set,] #
    tp <- length(data$Prediction[data[col] == type & data$class == type])
    fp <- length(data$Prediction[data[col] == type & data$class != type])
    fn <- length(data$Prediction[data[col] != type & data$class == type])
    tn <- length(data$Prediction[data[col] != type & data$class != type])
    precision <- tp / (tp + fp)
    recall <- tp / (tp + fn)
    f1 <- 2*(precision * recall) / (precision + recall)
    accuracy <- (tp) / length(data$Prediction[data$class == type])
    return(data.frame("class"=type,"reference"=ref,"method"=method,"size"=size,"set"=set,
                      "precision"=precision,"recall"=recall,"f1"=f1, "accuracy"=accuracy, bootstrap=col))
}

data_bootstrap$Size = 102
cols <- c("prediction_confidence", "prediction_confidenceFiltered", "prediction_binary", "prediction_filtered")
measures <- do.call(rbind, lapply(celltypes,         function(type) 
            do.call(rbind, lapply(methods,           function(method) 
            do.call(rbind, lapply(cols, function(col)
            do.call(rbind, lapply(unique(data$Set),  function(set) get_measures(data, type, "PBMC10x", method, 102,set, col)))))
                                                                                 
                                                                                
write.table(measures, "fast/AG_Haghverdi/Carla_Moelbert/Cell_Annotation/Results/Files/results_PBMCMosaic_qualitymeasures.csv",
            row.names=F, col.names=F, quote=F, sep=",")  
```
